```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imigrantes_canada.csv')
df
```

```python
df.info()
```

```python
df.set_index('País', inplace = True)
```

```python
# Intervalo de datas
intervalo = list(map(str, range(1980, 2014)))
```

```python
# Dados de interesse (Brasil)
df_br = pd.DataFrame(df.loc['Brasil', intervalo])
df_br.reset_index(inplace=True)
df_br.rename(columns={'index': 'ano', 'Brasil': 'imigrantes'},
             inplace=True)
df_br['imigrantes'] = df_br['imigrantes'].astype(int)

df_br
```

```python
min_ano = min(df_br['ano'].astype(int))
max_ano = max(df_br['ano'].astype(int))

plt.figure(figsize=(6, 3))
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.plot(df_br['ano'], df_br['imigrantes'])
plt.xticks([str(i) for i in range(min_ano, max_ano, 5)])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df_br['ano'], df_br['imigrantes'], lw=3)
ax.set_title('Imigração do Brasil para o Canadá (1980 à 2013)', fontsize=18,
             loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(df_br['ano'], df_br['imigrantes'])
ax[0].xaxis.set_major_locator(plt.MultipleLocator(4))
ax[0].set_title('Imigração do Brasil para o Canadá')
ax[0].set_xlabel('Ano')
ax[0].set_ylabel('Imigrantes')
ax[0].grid()

ax[1].boxplot(df_br['imigrantes'])
ax[1].set_xlabel('Brasil')
ax[1].grid()

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
         'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá - 1980 a 2013',
             loc='left', fontsize=16)
ax.set_ylabel('Número de imigrantes', fontsize=12)
plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
         'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá - 1980 a 2013',
             loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=12)
plt.show()
```

```python
america_sul_ordenado = america_sul.sort_values('Total')
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
         'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(america_sul_ordenado.index, america_sul_ordenado['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá - 1980 a 2013',
             loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=12)
plt.show()
```

```python
cores = []

for pais in america_sul_ordenado.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(america_sul_ordenado.index, america_sul_ordenado['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá - 1980 a 2013',
             loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=12)

for i, v in enumerate(america_sul_ordenado['Total']):
    ax.text(v + 10, i, str(v), color='black', fontsize=10, ha='left',
            va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```
