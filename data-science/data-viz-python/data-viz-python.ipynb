```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imigrantes_canada.csv')
df
```

```python
df.info()
```

```python
df.set_index('País', inplace = True)
```

```python
# Intervalo de datas
intervalo = list(map(str, range(1980, 2014)))
```

```python
# Dados de interesse (Brasil)
df_br = pd.DataFrame(df.loc['Brasil', intervalo])
df_br.reset_index(inplace = True)
df_br.rename(columns = {'index': 'ano', 'Brasil': 'imigrantes'},
             inplace = True)
df_br
```

```python
min_ano = min(df_br['ano'].astype(int))
max_ano = max(df_br['ano'].astype(int))

plt.figure(figsize = (6,3))
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.plot(df_br['ano'], df_br['imigrantes'])
plt.xticks([str(i) for i in range(min_ano, max_ano, 5)])
plt.show()
```
